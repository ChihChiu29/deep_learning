```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
def one_variable(x):
    return 7 * x + 4

x = np.random.rand(10000, 1)
y = one_variable(x)
```

```python
model = Sequential()
model.add(Dense(1, activation='linear', input_dim=1))
model.add(Dense(1, activation='linear'))
model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=1, batch_size=32)
```

```python
plot_x = np.linspace(start=-5, stop=5, num=100)

real_y = one_variable(plot_x)
model_y = model.predict(plot_x)

plt.plot(plot_x, real_y)
plt.plot(plot_x, model_y)
```

```python
def two_variables(x, y):
    return 7 * x - 2 * y + 4

x = np.random.rand(10000, 1)
y = np.random.rand(10000, 1)

train = np.column_stack((x, y))
z = two_variables(x, y)
```

```python
model = Sequential()
model.add(Dense(1, activation='linear', input_dim=2))
model.add(Dense(1, activation='linear'))
model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse')

model.fit(train, z, epochs=10, batch_size=32)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

plot_x = np.linspace(start=-5, stop=5, num=100)
plot_y = np.linspace(start=-5, stop=5, num=100)

mesh_x, mesh_y = np.meshgrid(plot_x, plot_y)

real_z = two_variables(mesh_x, mesh_y)

test = np.column_stack((mesh_x.reshape(10000, 1), mesh_y.reshape(10000, 1)))
model_z = model.predict(test)
model_z = model_z.reshape(100, 100)

ax.plot_surface(mesh_x, mesh_y, abs(real_z - model_z))
```

```python
def two_variables(x, y):
    return 7 * x ** 2 - 2 * y ** 2 + 4

x = np.random.rand(10000, 1)
y = np.random.rand(10000, 1)

train = np.column_stack((x, y))
z = two_variables(x, y)
```

```python
model = Sequential()
model.add(Dense(12, activation='relu', input_dim=2))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse')

model.fit(train, z, epochs=40, batch_size=32)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

plot_x = np.linspace(start=0, stop=1, num=100)
plot_y = np.linspace(start=0, stop=1, num=100)

mesh_x, mesh_y = np.meshgrid(plot_x, plot_y)

real_z = two_variables(mesh_x, mesh_y)

test = np.column_stack((mesh_x.reshape(10000, 1), mesh_y.reshape(10000, 1)))
model_z = model.predict(test)
model_z = model_z.reshape(100, 100)

ax.plot_surface(mesh_x, mesh_y, real_z)
ax.plot_surface(mesh_x, mesh_y, real_z - model_z)
```

```python
import numpy as np
x_train=np.random.rand(9000)
y_train=x_train**4+x_train**3-x_train
x_train=x_train.reshape(len(x_train),1)

x_test=np.linspace(0,1,100)
y_test=x_test**4+x_test**3-x_test
x_test=x_test.reshape(len(x_test),1)


model = Sequential()
model.add(Dense(units=200, input_dim=1))
model.add(Activation('relu'))
model.add(Dense(units=45))
model.add(Activation('relu'))
model.add(Dense(units=1))

model.compile(loss='mean_squared_error',
              optimizer='sgd')

model.fit(x_train, y_train, epochs=40, batch_size=50, verbose=1)

loss_and_metrics = model.evaluate(x_test, y_test, batch_size=100)

classes = model.predict(x_test, batch_size=1)

test=x_test.reshape(-1)
plt.plot(test,classes,c='r')
plt.plot(test,y_test,c='b')
plt.show()
```
