## Smooth non-linear function

```python
def one_variable(x):
    return 7 * x**2 - 9 * x + 4

x = np.random.rand(10000, 1)
y = one_variable(x)
```

```python
model = Sequential()
model.add(Dense(6, activation='relu', input_dim=1))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=40, batch_size=50)

plot_x = np.linspace(start=0, stop=1, num=100)

real_y = one_variable(plot_x)
model_y = model.predict(plot_x)

plt.plot(plot_x, real_y)
plt.plot(plot_x, model_y)
```

## Continuous indifferentiable function

```python
def one_variable(x):
    return (x < 0.5) * x + (x >= 0.5) * (1 - x)

x = np.random.rand(10000, 1)
y = one_variable(x)
```

```python
model = Sequential()
model.add(Dense(6, activation='relu', input_dim=1))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=40, batch_size=50)

plot_x = np.linspace(start=0, stop=1, num=100)

real_y = one_variable(plot_x)
model_y = model.predict(plot_x)

plt.plot(plot_x, real_y)
plt.plot(plot_x, model_y)
```

## Discontinous function

```python
def one_variable(x):
    return (x >= 0.5) * 2 - 1

x = np.random.rand(10000, 1)
y = one_variable(x)
```

```python
model = Sequential()
model.add(Dense(20, activation='relu', input_dim=1))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=40, batch_size=50)

plot_x = np.linspace(start=0, stop=1, num=100)

real_y = one_variable(plot_x)
model_y = model.predict(plot_x)

plt.plot(plot_x, real_y)
plt.plot(plot_x, model_y)
```
